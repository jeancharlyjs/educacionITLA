# Importaciones

```python
#Importaciones

import numpy as np
from sklearn.linear_model import LinearRegression
from random import Random
from pandas import DataFrame
import pandas as pl


import matplotlib.pyplot as plt
```

# Construir Modelo

```python
model = LinearRegression()
```

```python
ValorEntradas = np.random.normal(0.1, 1, 50)
ValorSalida = np.random.normal(0.1, 1, 50)
```

```python
df = DataFrame({
    "Años": [i for i in range(2000, 2050)],
    "Activos": ValorEntradas,
    "Gastos": ValorSalida
})
```

```python
df.describe()
```

```python
df[df["Activos"]>0.984675]
```

# Grafica

```python
plt.plot(df["Activos"], label="Activos")
plt.plot(df["Gastos"], label="Gastos" )
plt.legend()
plt.show()
```

```python
plt.scatter(df["Activos"], df["Gastos"])
```

```python
e1 = ValorEntradas.reshape(-1, 1)
s1 = ValorSalida.reshape(-1, 1)
```

```python
model.fit(e1, s1)
```

```python
model.coef_
```

```python
model.intercept_
```
