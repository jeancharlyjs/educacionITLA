````python
# Funcion Creada

def promedio(n1, n2, n3, n4) -> float:
    # Promedio = sum(valores)/ cantidad(valor)
    promedio = sum([n1, n2, n3, n4]) / 4
    # print("Promedio: ", promedio)
    return promedio

elemento1 = promedio()
````

````python
x = np.linspace(1, 5, 100000)
y = np.linspace(1, 5, 100000)


#Grafico
plt.title("Grafico de Funcioines Trigonometricas\n Estudiantes ITLA")
plt.plot(np.sin(x), label='Seno')
plt.plot(np.cos(y), label="Coseno")
plt.xlabel('sin')
plt.ylabel('cos')
plt.legend()
plt.show()
````

````python
def circunferencia(h, k, r, t):
    contar = 0
    for i in range(5):
        c = np.linspace(0, 2 * np.pi, t)
        x = h + r * np.cos(c)
        y = k + r * np.sin(c)
        contar += 1
        plt.plot(x / y * contar, y / y * x)
        
    return plt.show()

circunferencia(10, 10, 5, 100000)
````

````python
# Conjetura de Collatz

### Si N es par entonces Modulo 2
### Si N es impar entonces 3n + 1

def conjeturaCollatz(n, sec: list):
    sec.append(n)
    if n == 1:
        return 
    elif n % 2 == 0:
        return conjeturaCollatz(n / 2, sec)
    else:
        return conjeturaCollatz(3 * n + 1, sec)
````
