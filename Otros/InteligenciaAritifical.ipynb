# Neuronas Artificiales y su Funcionamiento

```python
class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.siguiente = None
class Enlace:
    def __init__(self):
        self.head = None
    def  agregar(self, dato):
        contar = 1

        nodoNuevo = Nodo(dato)
        if self.head is None:
            self.head = nodoNuevo
        else:
            cuncurrencia = self.head
            while cuncurrencia.siguiente is not None:
                print("Hola %s"%contar)
                cuncurrencia = cuncurrencia.siguiente
                contar += 1
            cuncurrencia.siguiente = nodoNuevo

    def cargarDatos(self):
        concurrencia = self.head
        while concurrencia is not None:
            print(concurrencia.dato)
            concurrencia = concurrencia.siguiente
```

```python
lista_ = Enlace()

lista_.agregar(1)
lista_.agregar(2)
lista_.agregar(3)
lista_.agregar(4)
```

# Sigmoide
## $$\alpha(x)= \frac{1}{1 + e^{-x}}$$

```python
def sigmoide(x):
    '''
    Esta funcion es para clasificar elemento binarios, en un rango de 0 a 1.
    '''
    return 1 / (1 + np.exp(-x))
```

```python
plt.figure(figsize=(7, 7))
plt.plot(sigmoide(np.arange(-10, 10, 0.1)))
plt.show()
```

# Tanh

$$ tanh(z) =\frac{e^z - e^{-z}}{e^z + e^{-z}}$$

```python
def tang(z):
    '''
    Esta función es de activación llamada Tangente Hiperbolica
    '''
    return ((np.exp(z) - np.exp(-z))) / ((np.exp(z) + np.exp(-z)))
plt.figure(figsize=(7, 7))
plt.plot(tang(np.arange(-10, 10, 0.1)))
plt.show()
```

# ReLU - (Unidad Lineal Rectificada)

$$ ReLU(x) = max(0, x)$$

```python
def reLU(x):
    '''
    Función de activacion que rectificadora
    '''
    return np.maximum(0, x)

plt.figure(figsize=(7, 7))
plt.plot(reLU(np.arange(-10, 10, 0.1)))
plt.show()
```

# ReLU - (Unidad Lineal Rectificada)

$$ LeakyReLU(x) = max(\alpha *  x, x)$$

```python
def leakyRelu(alph):
    def reLUActivacion(x):
        return np.maximum(alph * x, x)
    return reLUActivacion
plt.figure(figsize=(7, 7))
plt.plot(leakyRelu(0.1) (np.arange(-10, 10, 0.1)))
plt.show()
```
