$5x^3+4x^2-5$

````python
datos = np.linspace(-5, 5, 1000)

def funcion1(datos):
    return 5*datos**3+4*datos**2 -5
funcion1(datos)
````

````python
plt.plot(funcion1(datos))
plt.scatter(-5, funcion1(1))
````
