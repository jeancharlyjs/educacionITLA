````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def rosenbrock(x, y):
    """Funcin de Rosenbrock"""
    return (1 - x)**2 + 100 * (y - x**2)**2

def derivada_x(x, y):
    """Derivada parcial de la funcin de Rosenbrock respecto a x"""
    return -2 * (1 - x) - 400 * x * (y - x**2)

def derivada_y(x, y):
    """Derivada parcial de la funcin de Rosenbrock respecto a y"""
    return 200 * (y - x**2)

def gradiente_descendente(x, y, alpha=0.001, iteraciones=1000):
    """Algoritmo de descenso de gradiente"""
    historial_costo = []
    historial_x = []
    historial_y = []
    
    for i in range(iteraciones):
        # Calcular las derivadas parciales
        derivada_x_actual = derivada_x(x, y)
        derivada_y_actual = derivada_y(x, y)
        
        # Actualizar x y y usando el descenso de gradiente
        x -= alpha * derivada_x_actual
        y -= alpha * derivada_y_actual
        
        # Registrar el costo y las coordenadas actuales
        historial_costo.append(rosenbrock(x, y))
        historial_x.append(x)
        historial_y.append(y)
        
    return x, y, historial_costo, historial_x, historial_y

# Punto de inicio para el descenso de gradiente
x, y = -2, 2


# Realizar el descenso de gradiente
x_optimo, y_optimo, historial_costo, historial_x, historial_y = gradiente_descendente(x, y, alpha=0.001, iteraciones=1000)

# Crear una figura 3D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Crear una malla de puntos para la superficie de la funcin de Rosenbrock
x_vals = np.linspace(-3, 3, 100)
y_vals = np.linspace(-3, 3, 100)
x_mesh, y_mesh = np.meshgrid(x_vals, y_vals)
z_mesh = rosenbrock(x_mesh, y_mesh)

# Graficar la superficie de la funcin de Rosenbrock
ax.plot_surface(x_mesh, y_mesh, z_mesh, cmap='viridis', alpha=0.8)

# Graficar el historial de puntos encontrados por el descenso de gradiente
ax.plot(historial_x, historial_y, historial_costo, color='green', marker='o')


# Configuracin de la etiqueta de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Costo')

# Mostrar la grfica
plt.show()
````
