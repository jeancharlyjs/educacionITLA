## FunciÃ³n Rosenbrock

#### $z = f(x, y) = (a - x)^2+b(y - x^2)^2$

````python
#Importacion de librerias

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

````python
## Creacion de Funcion

def rosenbrock(x:float, y:float) -> float:
    return (1 - x) ** 2 + 100*(y - x**2)**2
````

````python
#Derivada de x

def derivacion_x(x:float, y:float) -> float:
    return -2 * (1 - x) - 400 * x * (y - x ** 2)

def derivacion_y(x:float, y:float) -> float:
    return  200 * (y - x ** 2)
````

````python
#Gradiante de la Fucnion Rosenbrock

def gradiante_desc_rosenbrock(x: float, y:float, alph=0.001, inter=10000):
    #Almacenamiento historico
    historia_costo = []
    historia_x = []
    historia_y = []

    ###Iteracion (Ejemplo 1)
    for i in range(inter):
        derivacion_x_actual = derivacion_x(x, y)
        derivacion_y_actual = derivacion_y(x, y)

        x -= alph * derivacion_x_actual
        y -= alph * derivacion_y_actual

        ###Agragar a la lista la salidas
        historia_costo.append(rosenbrock(x, y))
        historia_x.append(x)
        historia_y.append(y)
    return x, y, historia_costo, historia_x, historia_y

x = -1
y = 1

#Optimizadores

x_optimacion, y_optimacion, historico_costo, historico_x, historico_y = gradiante_desc_rosenbrock(x, y) #Ejemplo 2


    
````

### Sigmoide

# $\alpha(x) = \frac{1}{1 + e^{-x}}$

````python
import numpy as np
import matplotlib.pyplot as plt


def sigmoide(x):
    """
        Esta fucioin es creada para clasificar elementos binarios [perros, gatos], 
       [vegetacion sana, vegetacion enferma] en un rango de 0 a 1
    """
    return 1 / (1 + np.exp(-x))
````

````python
plt.figure(figsize=(5, 5))
plt.plot(sigmoide(np.arange(-10, 10, 0.1)))
plt.show()
````

## Tanh

# $tanh(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}}$

````python
def tanh(z):
    """
     Esta funcion es para activar llamada de Tangente Hiperbolicas
    """
    return ((np.exp(z) - np.exp(-z))) / ((np.exp(z) + np.exp(-z)))
````

````python
plt.figure(figsize=(5, 5))
plt.plot(tanh(np.arange(-10, 10, 0.1)))
plt.show()
````

# ReLU (Unidad Lineal Rectificada)
