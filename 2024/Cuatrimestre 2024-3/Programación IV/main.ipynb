# MVC - MODELO - VISTA - CONTROLADOR

````python
#Modelo
class Persona:
    """
        Esta clase es para definir a las personas que se 
        estaran registrado, donde se tomaran los siguiente
        atributos:
            - Nombre
            - Apellido
            - Edad
            - Sexo
            - Estado (Solo que el estado sera modificado mediante una 
                    metodo creado.)
    """
    def __init__(self, ids, nombre, apellido, edad, sexo):
        self.ids = ids
        self.nombre = nombre
        self.apellido = apellido
        self.edad = edad
        self.sexo = sexo
        self.estado = "Ocupado"
    
    def __str__(self):
        return "Bienvenido seas %s"%self.nombre

    def modificarEstado(self):
        """
            Esta fucion permite podificar el estado del atributo
            llamado <self.estado>, por lo que tambien se le 
            podra solicitar al objeto que seleccione el tipo
            de estado que en el cual se encuentre.
        """
        listaEstado = ['Comiendo', 'Estudiando', 'Durmiendo']
        for i in enumerate(listaEstado):
            print(i)
        resp = int(input("Seleccione la opcion deseada: "))
        self.estado = listaEstado[resp]
        return self.estado
````
