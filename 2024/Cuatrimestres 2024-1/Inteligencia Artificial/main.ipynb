````python
class Persona:
    def __init__(self, ids, nombre, apellido, edad, sexo, estatura, estado):
        self.ids = ids
        self.nombre = nombre
        self.apellido = apellido
        self.edad = edad
        self.sexo = sexo
        self.estatura = estatura
        self.estado = None

    def __repr__(self):
        return "Bienvenido %s %s"%(self.nombre, self.apellido)
    
    def modificador(self, n):
        if n is None:
            self.estado = None
        else:
            self.estado = n
            return self.estado
````

````python
p1 = Persona("234234", "Jean Charly", "Joseph",  33, False, 1.2, None)
p1.__repr__()
p1.modificador("Trabjando")
p1.estado
````

````python
p2 = Persona("2123123", "Nikola", "Tesla", 22, False, 2.1, "Trabjando")
p2.__repr__()
p2.modificador("Trabjando")
p2.estado
````

# Funcion de Rosenbrock


$$f(x, y) = (a - x)^2 + b(y - x^2)^2$$


````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

````python
#Crear la funcion de rosenbrock

def rosenbrock(x, y):
    return (1 - x) **2 + 100 * (y - x**2)**2


def derivada_x(x, y):
    return -2 * (1 - x) - 400 * x * (y - x**2)


def derivada_y(x, y):
    return 200 * (y - x**2)

def gradiente_desc(x, y, alpha=0.001, iteracion=1000):
    historial_costo = []
    historial_x = []
    historial_y = []

    for i in range(iteracion):
        derivada_x_actual = derivada_x(x, y)
        derivada_y_actual = derivada_y(x, y)

        x -= alpha * derivada_x_actual
        y -= alpha * derivada_y_actual

        historial_costo.append(rosenbrock(x, y))
        historial_x.append(x)
        historial_y.append(y)
    return x, y, historial_costo, historial_x, historial_y


x, y = -2, 2
x_optimacion, y_optimacion, historial_costo, historial_x, historial_y = gradiente_desc(x, y, alpha=0.001, iteracion=1000)


#Graficar
fig  = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

x_valor = np.linspace(-3, 3, 100)
y_valor = np.linspace(-3, 3, 100)
x_mesh, y_mesh = np.meshgrid(x_valor, y_valor)
z_mesh = rosenbrock(x_mesh, y_mesh)

ax.plot_surface(x_mesh, y_mesh, z_mesh, cmap='viridis', alpha=0.5)
ax.plot(historial_x, historial_y, historial_costo, color='blue', marker='o')



print(x_optimacion)
print(y_optimacion)
print(historial_costo)
print(min(historial_costo))
print(max(historial_costo))
````

````python
from sklearn.cluster import DBSCAN

# Cargar datos
datos = np.loadtxt("venta_propiedades.csv", delimiter=",")

# Identificar Clusters con diferentes distancias
distancias = ["euclidean", "chebyshev", "manhattan"]

fig = plt.figure(figsize=(12, 4))
plt.set_cmap("plasma")
for d in range(len(distancias)):
    clusters = DBSCAN(2, min_samples=10, 
                      metric=distancias[d]).fit_predict(datos)
    ax = fig.add_subplot(1, 3, d+1)
    # puntos considerados ruido
    ax.scatter(datos[clusters==-1][:, 0], datos[clusters==-1][:, 1], 
               c="cyan", s=200, marker="2")
    # puntos pertenencientes a clusters
    ax.scatter(datos[clusters>=0][:, 0], datos[clusters>=0][:, 1], 
               c=clusters[clusters>=0], s=200, 
               linewidth=0.5, edgecolors="black")
    ax.set_title(distancias[d].capitalize(), fontsize=24)
    ax.axis("off")
    
plt.show()
````
