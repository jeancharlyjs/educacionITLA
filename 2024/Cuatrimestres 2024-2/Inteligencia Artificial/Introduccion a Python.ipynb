### $\beta_0 = \frac{\sum^{n}_{i=1}(ingreso_i - mediaIngreso)(horas_i - mediaHora)}{\sum^{n}_{i=1}(ingreso_i - mediaIngreso)^2}$

````python
beta_0 = np.sum((df['ingreso'] - ingreso_media) * (df['horas'] - hora_media)) / np.sum((df['ingreso'] - ingreso_media)**2)
beta_0
````

### $\beta_1 = mediaHora -\beta_0*MediaIngreso$

````python
beta_1 = hora_media - beta_0 * ingreso_media
beta_1
````

### $hora_{pred}=\beta_0+\beta_1 * ingreso_{nuevo}$

````python
valorNuevo = 9973.577534
hora_pred = beta_1 + beta_0 * valorNuevo
hora_pred
````

#### $z = f(x, y) = (a - x)^2 + b(y - x^2)^2$

````python
#Creacion de funcion  Rosenbrok

#Funcion rosenbrok
def rosenbrok(x, y):
    return (1 - x) ** 2 + 100 * (y - x**2)**2


#Derivada de x

def derivada_x(x, y):
    return -2 * (1 - x) - 400 * x * (y - x **2)

def derivada_y(x, y):
    return 200 * (y - x** 2)

def gradiente_desc(x, y, alph=0.001, iteracion=10000):
    historico_costo = []
    historico_x = []
    historico_y = []

    for i in range(iteracion):
        derivada_x_actual = derivada_x(x, y)
        derivada_y_actual = derivada_y(x, y)

        x -= alph * derivada_x_actual
        y -= alph * derivada_y_actual

        historico_costo.append(rosenbrok(x, y))
        historico_x.append(x)
        historico_y.append(y)
    return x, y, historico_costo, historico_x, historico_y

x, y = -1, 1
x_optimacion, y_optimaacion, historico_costo, historico_x, historico_y = gradiente_desc(x, y)
````

### Centroide

### $c = (\frac{x_1 + x_2 + x_3 + \ldots+{x_n}}{k}, \frac{y_1 + y_2 + y_3 + \ldots+{y_n}}{k})$

### $c = (\frac{\sum_{i=1}^{n} x_i}{n}, \frac{\sum_{i=1}^{n} y_i}{n})$

````python
df.head()
````

````python
from sklearn.cluster import DBSCAN

dist = ['euclidean', 'manhattan', 'chebyshev']
fig = plt.figure(figsize=(15, 5))
plt.set_cmap('plasma')

for i in range(len(dist)):
    c = DBSCAN(2, min_samples=10, metric=dist[i]).fit_predict(df)
    ax = fig.add_subplot(1, 3, i+1)
    ax.scatter(df[c==-1]['metro'], df[c==-1]['costo'], c='black', alpha=0.2) #No pertenecen
    ax.scatter(df[c>=0]['metro'], df[c>=0]['costo'], c=c[c>=0], s=150, alpha=0.5) #Si pertenecen

    ax.set_title(dist[i].capitalize(), fontsize=24)
````

````python
# Paso 1: Importar librerías
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

# Paso 2: Cargar datos
data = load_iris()
X = data.data          # Características (largo y ancho de pétalos y sépalos)
y = data.target        # Etiquetas (setosa, versicolor, virginica)

# Paso 3: Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Paso 4: Entrenar modelo Naive Bayes (usa probabilidad condicional)
model = GaussianNB()
model.fit(X_train, y_train)

# Paso 5: Predecir
y_pred = model.predict(X_test)

# Paso 6: Evaluación (esperanza matemática como precisión esperada)
print("Precisión esperada del modelo:", accuracy_score(y_test, y_pred))
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred, target_names=data.target_names))

# Paso 7: Mostrar algunas probabilidades condicionales
print("\nProbabilidades condicionales para la primera muestra de test:")
probs = model.predict_proba([X_test[0]])
for clase, prob in zip(data.target_names, probs[0]):
    print(f"{clase}: {prob:.2f}")
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# 1. Cargar datos Iris
iris = load_iris()
X = iris.data
y = iris.target
class_names = iris.target_names

# 2. Separar en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# 3. Entrenar el modelo Naive Bayes
model = GaussianNB()
model.fit(X_train, y_train)

# 4. Predecir
y_pred = model.predict(X_test)

# 5. Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)

# 6. Calcular las probabilidades condicionales para la primera muestra
first_sample = X_test[0].reshape(1, -1)
probs = model.predict_proba(first_sample).flatten()

# 7. Graficar
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Matriz de Confusión
disp.plot(ax=axs[0], cmap='Blues', colorbar=True)
axs[0].set_title("Matriz de Confusión")

# Probabilidades Condicionales
axs[1].bar(class_names, probs, color='coral')
axs[1].set_title("Probabilidades condicionales de la primera muestra")
axs[1].set_ylabel("Probabilidad")
axs[1].set_ylim([0, 1.0])

plt.tight_layout()
plt.show()
````
