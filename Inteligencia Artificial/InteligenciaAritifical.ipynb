# Sigmoide
## $$\alpha(x)= \frac{1}{1 + e^{-x}}$$


```python
def sigmoide(x):
    '''
    Esta funcion es para clasificar elemento binarios, en un rango de 0 a 1.
    '''
    return 1 / (1 + np.exp(-x))
```

```python
plt.figure(figsize=(7, 7))
plt.plot(sigmoide(np.arange(-10, 10, 0.1)))
plt.show()
```

# Tanh

$$ tanh(z) =\frac{e^z - e^{-z}}{e^z + e^{-z}}$$

```python
def tang(z):
    '''
    Esta función es de activación llamada Tangente Hiperbolica
    '''
    return ((np.exp(z) - np.exp(-z))) / ((np.exp(z) + np.exp(-z)))
plt.figure(figsize=(7, 7))
plt.plot(tang(np.arange(-10, 10, 0.1)))
plt.show()
```

# ReLU - (Unidad Lineal Rectificada)

$$ ReLU(x) = max(0, x)$$

```python
def reLU(x):
    '''
    Función de activacion que rectificadora
    '''
    return np.maximum(0, x)

plt.figure(figsize=(7, 7))
plt.plot(reLU(np.arange(-10, 10, 0.1)))
plt.show()
```

# ReLU - (Unidad Lineal Rectificada)

$$ LeakyReLU(x) = max(\alpha *  x, x)$$

```python
def leakyRelu(alph):
    def reLUActivacion(x):
        return np.maximum(alph * x, x)
    return reLUActivacion
plt.figure(figsize=(7, 7))
plt.plot(leakyRelu(0.1) (np.arange(-10, 10, 0.1)))
plt.show()
```

```python
# Vamos a definir una funcion que evalue el coste

# Error Cuadratico
def coste(x, y, theta0, theta1): #theta's
    m = len(x)
    hipotesis = theta0 + theta1 * x
    error = hipotesis - y
    return np.sum(error ** 2) / (2 * m)

# Definicion de las Derivadas Parciales en funcion del Coste
def derivada_theta0(x, y):
    return lambda theta0, theta1: np.sum(theta0 + theta1 * x - y) / len(x)

def derivada_theta1(x, y):
    print(lambda theta0, theta1: np.sum((theta0 + theta1 * x - y) * x)/ len(x))
    return lambda theta0, theta1: np.sum((theta0 + theta1 * x - y) * x)/ len(x)

def gradiante_decendente(x, y, theta0=0, theta1=1, alpha=0.01, iteracion=1000):
    historico_coste = []
    t = []
    t1 = []
    for i in range(iteracion):
        theta0_deriv = derivada_theta0(x, y)(theta0, theta1)
        theta1_deriv = derivada_theta1(x, y)(theta0, theta1)
        theta0_deriv -= alpha * theta0_deriv
        theta1_deriv -= alpha * theta1_deriv
        historico_coste.append(coste(x, y, theta0, theta1))
        t.append(theta0_deriv)
        t1.append(theta1_deriv)
        return theta0, theta1, historico_coste, t, t1
        
```

# Función de Rosenbrock


```python
import numpy as np
import matplotlib.pyplot as plt
```

$f(\mathbf {x} )=\sum _{i=1}^{N-1}100(x_{i+1}-x_{i}^{2})^{2}+(1-x_{i})^{2}\quad $

```python
def ros(x, y):
    return (1 - x)** 2 + 100 *(y - x ** x) ** 2

def derivada_x(x, y):
    return -2 * (1 - x) - 400 * x * (y - x ** 2)

def derivada_y(x, y):
    return 200 * (y - x ** 2)

def gradiante_decendente(x, y, alpha=0.001, iteracion=1000):
    hist_coste = []
    hist_x = []
    hist_y = []
    for i in range(iteracion):
        x_derivada = derivada_x(x, y)
        y_derivada = derivada_y(x, y)
        x -= alpha * x_derivada
        y -= alpha * y_derivada
        hist_coste.append(ros(x, y))
        hist_x.append(x)
        hist_y.append(y)
    return x, y, hist_coste, hist_x, hist_y
```

```python
## Valores

x, y = -2, 2
x_opt, y_opt, h, h_x, h_y = gradiante_decendente(x, y, alpha=0.001, iteracion=1000)
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
x_valor = np.linspace(-3, 3, 100)
y_valor = np.linspace(-3, 3, 100)
m_x, m_y = np.meshgrid(x_valor, y_valor)
m_z = ros(m_y, m_x)
ax.plot_surface(m_x, m_y, m_z, cmap='viridis', alpha=0.8)
ax.plot(h_x, h_y, h, color='red', marker='o')
```
